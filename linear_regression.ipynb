# Linear Regression Visualization

```python
import matplotlib.pyplot as plt
import  numpy as np
import pandas as pd
```

```python
datafr= {
    "sizes": [340,200,450,540,220,665,322,125,875,434,232,654],
    "prices":[1200,1300,1100,1267,1245,1675,1566,1355,1899,1345,1566,1900]
}
```

```python
pd.DataFrame(data=datafr)
```

```python
x_train= list(datafr["sizes"])
y_train= list(datafr["prices"])
```

```python
plt.scatter(x_train, y_train, marker="^", c="y", label="Data Points")
plt.show()
```

```python
w= 3
b=3.5
```

```python
def compute_hypothesis(x, w,b):

    m= len(x)

    f_model=np.zeros(m)

    for i in range(m):
        f_model[i]= w*x[i]+b

    return f_model
```

```python

model_predictions= compute_hypothesis(x_train,w,b)
plt.plot(x_train, model_predictions, c='r', label= "fitted line")
plt.scatter(x_train, y_train, marker= "^",c="g", label= "Sample Data Points")
plt.xlabel("House Sizes")
plt.ylabel("House Prices")
plt.title("House rent prediction model(Linear regression)")
plt.show()
```
